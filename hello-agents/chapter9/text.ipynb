```python
from hello_agents import SimpleAgent, HelloAgentsLLM, ToolRegistry
from hello_agents.context import ContextBuilder, ContextConfig
from hello_agents.tools import MemoryTool, RAGTool

class ContextAwareAgent(SimpleAgent):
    """å…·æœ‰ä¸Šä¸‹æ–‡æ„ŸçŸ¥èƒ½åŠ›çš„ Agent"""

    def __init__(self, name: str, llm: HelloAgentsLLM, **kwargs):
        super().__init__(name=name, llm=llm, system_prompt=kwargs.get("system_prompt", ""))

        # åˆå§‹åŒ–ä¸Šä¸‹æ–‡æ„å»ºå™¨
        self.memory_tool = MemoryTool(user_id=kwargs.get("user_id", "default"))
        self.rag_tool = RAGTool(knowledge_base_path=kwargs.get("knowledge_base_path", "./kb"))

        self.context_builder = ContextBuilder(
            memory_tool=self.memory_tool,
            rag_tool=self.rag_tool,
            config=ContextConfig(max_tokens=4000)
        )

        self.conversation_history = []

    def run(self, user_input: str) -> str:
        """è¿è¡Œ Agent,è‡ªåŠ¨æ„å»ºä¼˜åŒ–çš„ä¸Šä¸‹æ–‡"""

        # 1. ä½¿ç”¨ ContextBuilder æ„å»ºä¼˜åŒ–çš„ä¸Šä¸‹æ–‡
        optimized_context = self.context_builder.build(
            user_query=user_input,
            conversation_history=self.conversation_history,
            system_instructions=self.system_prompt
        )

        # 2. ä½¿ç”¨ä¼˜åŒ–åçš„ä¸Šä¸‹æ–‡è°ƒç”¨ LLM
        messages = [
            {"role": "system", "content": optimized_context},
            {"role": "user", "content": user_input}
        ]
        response = self.llm.invoke(messages)

        # 3. æ›´æ–°å¯¹è¯å†å²
        from hello_agents.core.message import Message
        from datetime import datetime

        self.conversation_history.append(
            Message(content=user_input, role="user", timestamp=datetime.now())
        )
        self.conversation_history.append(
            Message(content=response, role="assistant", timestamp=datetime.now())
        )

        # 4. å°†é‡è¦äº¤äº’è®°å½•åˆ°è®°å¿†ç³»ç»Ÿ
        self.memory_tool.execute(
            "add",
            content=f"Q: {user_input}\nA: {response[:200]}...",  # æ‘˜è¦
            memory_type="episodic",
            importance=0.6
        )

        return response
```

```python
# ä½¿ç”¨ç¤ºä¾‹
agent = ContextAwareAgent(
    name="æ•°æ®åˆ†æé¡¾é—®",
    llm=HelloAgentsLLM(),
    system_prompt="ä½ æ˜¯ä¸€ä½èµ„æ·±çš„Pythonæ•°æ®å·¥ç¨‹é¡¾é—®ã€‚",
    user_id="user123",
    knowledge_base_path="./data_science_kb"
)

response = agent.run("å¦‚ä½•ä¼˜åŒ–Pandasçš„å†…å­˜å ç”¨?")
print(response)
```

```python
import pandas as pd
import numpy as np

# è¯»å– Excel æ–‡ä»¶
df = pd.read_excel("/mnt/d/studying-code/Large-Model-Application-Development/hello-agents/chapter9/å‰¯æœ¬sta-åŠ¨ç”»æµç•…ä¸è¿è´¯æ€§.xlsx")

# å¦‚æœç¬¬ä¸€åˆ—æ˜¯â€œåºå·â€ï¼Œé€šå¸¸ä¸å‚ä¸çŸ©é˜µæ•°æ®ï¼Œå¯ä»¥åˆ æ‰
if df.columns[0] in ["åºå·", "ç¼–å·", "ID"]:
    df = df.iloc[:, 1:]   # åˆ é™¤ç¬¬ä¸€åˆ—

# å°† dataframe è½¬æ¢æˆ numpy çŸ©é˜µ
matrix = df.to_numpy()

print("çŸ©é˜µå½¢çŠ¶ï¼ˆç”¨æˆ·æ•° Ã— é¢˜é¡¹æ•°ï¼‰:", matrix.shape)
print(matrix)
```

```python
import numpy as np
from scipy import stats

# participants_data = matrix  # ä½ ä¹‹å‰è¯»å…¥çš„çŸ©é˜µ
participants_data = matrix

# --- 2. Calculate Preference Scores ---
user_scores = []
for p_trials in participants_data:

    # ä½¿ç”¨ numpy ç»Ÿè®¡ 1 å’Œ -1 çš„æ•°é‡
    wins = np.sum(p_trials == 1)
    losses = np.sum(p_trials == -1)
    total = len(p_trials)

    # Calculate normalized score (-1 to 1)
    score = (wins - losses) / total
    user_scores.append(score)

print(f"User Scores: {user_scores}")

# --- 3. Descriptive Stats (Median & IQR) ---
median = np.median(user_scores)
q75, q25 = np.percentile(user_scores, [75 ,25])
iqr = q75 - q25

print("\n--- Descriptive Statistics ---")
print(f"Median (Mdn): {median:.4f}")
print(f"IQR: {iqr:.4f}")

# --- 4. The Statistical Test (Wilcoxon) ---
w_stat, p_value = stats.wilcoxon(user_scores)

# Approximate Z-score for reporting
z_score = stats.norm.ppf(p_value/2) * -1  # two-tailed

# Effect Size
N = len(user_scores)
r_effect = abs(z_score) / np.sqrt(N)

print("\n--- Inference Statistics ---")
print(f"Wilcoxon Statistic (W): {w_stat:.4f}")
print(f"P-value: {p_value:.5f}")
print(f"Z-score (approx): {z_score:.4f}")
print(f"Effect Size (r): {r_effect:.4f}")
```
