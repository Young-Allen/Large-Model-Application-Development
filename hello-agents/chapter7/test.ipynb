```python
# test_my_calculator.py
from dotenv import load_dotenv
from my_calculator_tool import create_calculator_registry

# Âä†ËΩΩÁéØÂ¢ÉÂèòÈáè
load_dotenv()

def test_calculator_tool():
    """ÊµãËØïËá™ÂÆö‰πâËÆ°ÁÆóÂô®Â∑•ÂÖ∑"""

    # ÂàõÂª∫ÂåÖÂê´ËÆ°ÁÆóÂô®ÁöÑÊ≥®ÂÜåË°®
    registry = create_calculator_registry()

    print("üß™ ÊµãËØïËá™ÂÆö‰πâËÆ°ÁÆóÂô®Â∑•ÂÖ∑\n")

    # ÁÆÄÂçïÊµãËØïÁî®‰æã
    test_cases = [
        "2 + 3",           # Âü∫Êú¨Âä†Ê≥ï
        "10 - 4",          # Âü∫Êú¨ÂáèÊ≥ï
        "5 * 6",           # Âü∫Êú¨‰πòÊ≥ï
        "15 / 3",          # Âü∫Êú¨Èô§Ê≥ï
        "sqrt(16)",        # Âπ≥ÊñπÊ†π
    ]

    for i, expression in enumerate(test_cases, 1):
        print(f"ÊµãËØï {i}: {expression}")
        result = registry.execute_tool("my_calculator", expression)
        print(f"ÁªìÊûú: {result}\n")

def test_with_simple_agent():
    """ÊµãËØï‰∏éSimpleAgentÁöÑÈõÜÊàê"""
    from hello_agents import HelloAgentsLLM

    # ÂàõÂª∫LLMÂÆ¢Êà∑Á´Ø
    llm = HelloAgentsLLM()

    # ÂàõÂª∫ÂåÖÂê´ËÆ°ÁÆóÂô®ÁöÑÊ≥®ÂÜåË°®
    registry = create_calculator_registry()

    print("ü§ñ ‰∏éSimpleAgentÈõÜÊàêÊµãËØï:")

    # Ê®°ÊãüSimpleAgent‰ΩøÁî®Â∑•ÂÖ∑ÁöÑÂú∫ÊôØ
    user_question = "ËØ∑Â∏ÆÊàëËÆ°ÁÆó sqrt(16) + 2 * 3"

    print(f"Áî®Êà∑ÈóÆÈ¢ò: {user_question}")

    # ‰ΩøÁî®Â∑•ÂÖ∑ËÆ°ÁÆó
    calc_result = registry.execute_tool("my_calculator", "sqrt(16) + 2 * 3")
    print(f"ËÆ°ÁÆóÁªìÊûú: {calc_result}")

    # ÊûÑÂª∫ÊúÄÁªàÂõûÁ≠î
    final_messages = [
        {"role": "user", "content": f"ËÆ°ÁÆóÁªìÊûúÊòØ {calc_result}ÔºåËØ∑Áî®Ëá™ÁÑ∂ËØ≠Ë®ÄÂõûÁ≠îÁî®Êà∑ÁöÑÈóÆÈ¢ò:{user_question}"}
    ]

    print("üìù ÊûÑÂª∫ÊúÄÁªàÊ∂àÊÅØ:", final_messages)
    
    # print("\nüéØ SimpleAgentÁöÑÂõûÁ≠î:")
    # response = llm.think(final_messages)
    # for chunk in response:
    #     print(chunk, end="", flush=True)
    # print("\n")
    full = "".join(llm.think(final_messages))
    # ‰∏çÂÜçËæπÊî∂Ëæπ printÔºåÂè™Âú®ÊúÄÂêéÊâìÂç∞‰∏ÄÊ¨°Ôºö
    print("\nüéØ SimpleAgentÁöÑÂõûÁ≠î(ÂÆåÊï¥):\n" + full)
```
